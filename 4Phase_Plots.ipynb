```python
import SNIDsn
import SNIDdataset as snid
import numpy as np
import SNePCA

import plotly.plotly as ply
import plotly.graph_objs as go
import plotly.tools as tls

import matplotlib.pyplot as plt
```

```python
dataset0 = snid.loadDataset(pathdir='allSNIDtemp/', snlist='snlist.txt')
dataset5 = snid.loadDataset(pathdir='allSNIDtemp/', snlist='snlist.txt')
dataset10 = snid.loadDataset(pathdir='allSNIDtemp/', snlist='snlist.txt')
dataset15 = snid.loadDataset(pathdir='allSNIDtemp/', snlist='snlist.txt')

dataset20 = snid.loadDataset(pathdir='allSNIDtemp/', snlist='snlist.txt')
dataset25 = snid.loadDataset(pathdir='allSNIDtemp/', snlist='snlist.txt')
```

```python
typeinfo = snid.datasetTypeDict(dataset0)
sesn = np.concatenate((typeinfo['Ib'], typeinfo['IIb'], typeinfo['Ic'], typeinfo['IcBL']))
dataset0 = snid.subset(dataset0, sesn)

typeinfo = snid.datasetTypeDict(dataset5)
sesn = np.concatenate((typeinfo['Ib'], typeinfo['IIb'], typeinfo['Ic'], typeinfo['IcBL']))
dataset5 = snid.subset(dataset5, sesn)

typeinfo = snid.datasetTypeDict(dataset10)
sesn = np.concatenate((typeinfo['Ib'], typeinfo['IIb'], typeinfo['Ic'], typeinfo['IcBL']))
dataset10 = snid.subset(dataset10, sesn)

typeinfo = snid.datasetTypeDict(dataset15)
sesn = np.concatenate((typeinfo['Ib'], typeinfo['IIb'], typeinfo['Ic'], typeinfo['IcBL']))
dataset15 = snid.subset(dataset15, sesn)




typeinfo = snid.datasetTypeDict(dataset20)
sesn = np.concatenate((typeinfo['Ib'], typeinfo['IIb'], typeinfo['Ic'], typeinfo['IcBL']))
dataset20 = snid.subset(dataset20, sesn)

typeinfo = snid.datasetTypeDict(dataset25)
sesn = np.concatenate((typeinfo['Ib'], typeinfo['IIb'], typeinfo['Ic'], typeinfo['IcBL']))
dataset25 = snid.subset(dataset25, sesn)
```

```python
snid.snidsetNAN(dataset0)
snid.snidsetNAN(dataset5)
snid.snidsetNAN(dataset10)
snid.snidsetNAN(dataset15)

snid.snidsetNAN(dataset20)
snid.snidsetNAN(dataset25)
```

```python
snid.interpGaps(dataset0, 4000, 7000, 20)
snid.interpGaps(dataset5, 4000, 7000, 20)
snid.interpGaps(dataset10, 4000, 7000, 20)
snid.interpGaps(dataset15, 4000, 7000, 20)

snid.interpGaps(dataset20, 4000, 7000, 20)
snid.interpGaps(dataset25, 4000, 7000, 20)
```

```python
snid.smoothSpectra(dataset0, velcut=1000, velcutIcBL=3000, plot=False)
snid.smoothSpectra(dataset5, velcut=1000, velcutIcBL=3000, plot=False)
snid.smoothSpectra(dataset10, velcut=1000, velcutIcBL=3000, plot=False)
snid.smoothSpectra(dataset15, velcut=1000, velcutIcBL=3000, plot=False)

snid.smoothSpectra(dataset20, velcut=1000, velcutIcBL=3000, plot=False)
snid.smoothSpectra(dataset25, velcut=1000, velcutIcBL=3000, plot=False)
```

# Eigenspectra

```python
f, axs = plt.subplots(5,1,figsize=(10,20))
```

```python
for i in range(5):
    ax = axs[i]
    ax.set_title('PCA%d Eigenspectra'%(i+1))
    ax.set_xlim((3800,7700))
    ax.set_yticks([])
    ax.plot(snidPCA0.wavelengths,snidPCA0.evecs[i]+2-0,'r',label='phase 0')
    ax.plot(snidPCA0.wavelengths,snidPCA5.evecs[i]+2-.5,'b',label='phase 5')
    ax.plot(snidPCA0.wavelengths,snidPCA10.evecs[i]+2-1,'c',label='phase 10')
    ax.plot(snidPCA0.wavelengths,snidPCA15.evecs[i]+2-1.5,'g',label='phase 15')
    
    ax.plot(snidPCA20.wavelengths,snidPCA20.evecs[i]+2-2,'k',label='phase 20')
    ax.plot(snidPCA25.wavelengths,snidPCA25.evecs[i]+2-2.5,'y',label='phase 25')
    
    ax.legend()
```
